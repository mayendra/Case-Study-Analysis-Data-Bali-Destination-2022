Bali Popular Destination for Tourist 2022

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Bali Popular Destination for Tourist 2022 - Sheet1.csv")
df
```

Drop several coulumn

```python
clean_data = df.drop(columns=['Location','Coordinate','Source','Description','Tourism/Visitor Fee (approx in USD)'])
clean_data
```

Sorting by Google Maps Rating


```python
clean_data.sort_values(by='Google Maps Rating',ascending=False)
clean_data.reset_index(drop=True)
```

```python
jumlah_rating = clean_data["Google Maps Rating"].value_counts()
jumlah_rating
```

```python
plt.figure(figsize=(8, 6))
plt.pie(jumlah_rating, labels=jumlah_rating.index, autopct='%1.1f%%', startangle=140)
plt.title('Total Rating Google Maps Bali Destination')
plt.axis('equal')
plt.show()
```
